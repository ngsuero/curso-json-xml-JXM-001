# Estructura de un Documento XML

XML organiza informaci√≥n en un **√°rbol** de nodos.

## Elementos y contenido

Un **elemento** tiene:
- Nombre de etiqueta
- Opcionalmente atributos
    
Ejemplo:

````typescript
const xml = `
<producto>
  <nombre>Teclado</nombre>
  <precio>79.95</precio>
</producto>`;
console.log(xml);
````

Elementos pueden tener **texto, atributos u otros elementos** dentro:

````typescript
const xml2 = `
<producto id="p1001">
  <nombre>Pantalla</nombre>
  <descripcion>24 pulgadas</descripcion>
</producto>`;
console.log(xml2);
````

## Texto vs Elementos

XML diferencia entre **texto** y **estructura**.

Ejemplo con texto directo:

````typescript
const xmlTexto = `<mensaje>Hola Mundo</mensaje>`;
console.log(xmlTexto);
````

Ejemplo con elementos dentro:

````typescript
const xmlConElementos = `
<mensaje>
  <texto>Hola</texto>
  <origen>Soporte</origen>
</mensaje>`;
console.log(xmlConElementos);
````

Informaci√≥n adicional que acompa√±a al elemento:
