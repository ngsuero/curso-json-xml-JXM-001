Cuando dos sistemas diferentes env√≠an XML, pueden usar **las mismas etiquetas con distinto significado**.

## ¬øQu√© es un Namespace?

Es una forma de **identificar** de d√≥nde proviene un conjunto de etiquetas.

Se define como un espacio √∫nico que se representa con un **URI**, por ejemplo:
- `http://ejemplo.com/productos`

## Usando prefijos

Asociamos un **prefijo** a un namespace:


````typescript
const xmlNS = `
<prod:producto xmlns:prod="http://ejemplo.com/productos">
  <prod:nombre>Teclado</prod:nombre>
  <prod:precio>59.99</prod:precio>
</prod:producto>`;

console.log(xmlNS);
````

El **prefijo** `prod:` dice que estas etiquetas pertenecen al namespace
`http://ejemplo.com/productos`

## Default Namespace

Si todos los elementos pertenecen al mismo namespace, podemos definir uno **por defecto**:

````typescript
const xmlDefaultNS = `
<catalogo xmlns="http://ejemplo.com/catalogo">
  <producto>
    <nombre>Raton</nombre>
    <precio>20</precio>
  </producto>
</catalogo>`;
console.log(xmlDefaultNS);
````

## Namespaces M√∫ltiples

Cuando mezclamos datos de **fuentes distintas**:

````typescript
const xmlMulti = `
<pedido xmlns:prod="http://ejemplo.com/productos" xmlns:user="http://ejemplo.com/usuarios">
  <user:cliente>Juan</user:cliente>
  <prod:producto>Teclado</prod:producto>
</pedido>`;
console.log(xmlMulti);
````
