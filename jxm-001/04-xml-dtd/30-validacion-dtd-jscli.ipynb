Ejemplo de XML con DTD **interna**:

````typescript
const xml = `\
<!DOCTYPE catalogo [
  <!ELEMENT catalogo (producto+)>
  <!ELEMENT producto (nombre, precio)>
  <!ELEMENT nombre (#PCDATA)>
  <!ELEMENT precio (#PCDATA)>
]>
<catalogo>
  <producto>
    <nombre>Teclado</nombre>
    <precio>80</precio>
  </producto>
</catalogo>`;

console.log(xml);
````

üëâ Validaci√≥n con `xmllint-wasm`:

````typescript
import init, { validateXML } from "https://esm.run/xmllint-wasm";
await init();

const result = validateXML({ xml });
console.log(result.valid ? "XML v√°lido ‚úî" : result.errors);
````
