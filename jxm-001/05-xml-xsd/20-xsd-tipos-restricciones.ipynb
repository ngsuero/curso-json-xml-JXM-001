## Restricciones de texto
Ejemplo: limitar longitud del nombre del producto

````typescript
const xsdNombre = `
<xs:element name="nombre">
  <xs:simpleType>
    <xs:restriction base="xs:string">
      <xs:minLength value="3" />
      <xs:maxLength value="50" />
    </xs:restriction>
  </xs:simpleType>
</xs:element>`;
console.log(xsdNombre);
````

## Enumeraciones
Sirven para limitar los valores posibles:

````typescript
const xsdEstado = `
<xs:element name="estado">
  <xs:simpleType>
    <xs:restriction base="xs:string">
      <xs:enumeration value="nuevo" />
      <xs:enumeration value="reaco" />
      <xs:enumeration value="usado" />
    </xs:restriction>
  </xs:simpleType>
</xs:element>`;
console.log(xsdEstado);
````
