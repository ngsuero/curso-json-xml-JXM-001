````typescript
import avro from "npm:avro-js";

const schema = {
  type: "record",
  name: "Person",
  fields: [
    { name: "id", type: "int" },
    { name: "name", type: "string" },
    { name: "age", type: "int" }
  ]
};

const PersonType = avro.parse(schema);

const person = { id: 1, name: "Rick Sanchez", age: 70 };

const bytes = PersonType.toBuffer(person);
console.log("Bytes:", bytes);
````

````typescript
const decoded = PersonType.fromBuffer(bytes);
console.log("Decoded:", decoded);
````
