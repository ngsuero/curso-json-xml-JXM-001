# Sintaxis y estructuras JSON

JSON es un formato de **texto plano** usado para transmitir datos estructurados en la web.
‚ö†Ô∏è Aunque JavaScript lo entiende bien, **JSON NO es JavaScript**.

## Tipos v√°lidos en JSON

| Tipo | Ejemplo |
|------|---------|
| string | "Hola" |
| number | 42 |
| boolean | true / false |
| null | null |
| object | { "a": 1 } |
| array | [1,2,3] |

‚ùå JSON **NO admite**: comentarios, funciones, undefined

## Objetos JSON
- Llaves `{}`
- Claves SIEMPRE entre comillas dobles
- Valores: cualquier tipo v√°lido

````typescript
// Ejemplo de objeto JSON correcto
const producto = `{
  "nombre": "Raton",
  "precio": 20,
  "stock": 10
}`;
console.log(producto);
````
