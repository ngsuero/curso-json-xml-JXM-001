## JSON.parse()
Convierte una cadena JSON en un **objeto de JavaScript**.

````typescript
const texto = `{"nombre": "Raton", "precio": 20}`;
const obj = JSON.parse(texto);
console.log(obj);
console.log(obj.nombre);
````

## JSON.stringify()
Convierte un objeto JS en **texto JSON**.
**Ideal para enviar datos a API o guardar en LocalStorage.**

````typescript
const usuario = { email: "ana@example.com", activo: true };
const json = JSON.stringify(usuario);
console.log(json);
````
